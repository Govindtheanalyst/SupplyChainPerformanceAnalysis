# **Data Cleaning**

```python
#Dropping Unnecessary columns
cols_to_drop = [
    "Product Description",
    "Product Image",
    "Customer Email",
    "Customer Password",
    "Customer Fname",
    "Customer Lname",
    "Customer Street",
    "Customer Zipcode",
    "Order Zipcode"
]

df = df.drop(columns=cols_to_drop)

df.shape
```

```python
#Standardising the COlumns by converting them into snake_case
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_", regex=False)
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
    .str.replace("/", "_", regex=False)
    .str.replace("-", "_", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace(".", "", regex=False)
)
```

```python
#verifying the data
df.columns.tolist()
```

```python
#converting orderdate and shipdate to datetime
df["order_date_dateorders"] = pd.to_datetime(df["order_date_dateorders"])
df["shipping_date_dateorders"] = pd.to_datetime(df["shipping_date_dateorders"])
```

```python
#verifying
df[["order_date_dateorders", "shipping_date_dateorders"]].dtypes
```

# **Feature Engineering**

```python
df["order_year"] = df["order_date_dateorders"].dt.year
df["order_month"] = df["order_date_dateorders"].dt.month
df["order_month_name"] = df["order_date_dateorders"].dt.month_name()
df["order_day"] = df["order_date_dateorders"].dt.day
df["order_day_name"] = df["order_date_dateorders"].dt.day_name()
df["order_quarter"] = df["order_date_dateorders"].dt.quarter
```

```python
#Creating shipping delay
df["shipping_delay"] = (
    df["days_for_shipping_real"] -
    df["days_for_shipment_scheduled"]
)
```

```python
#Creating a readable status for better visualisation
df["late_delivery"] = df["late_delivery_risk"].map({
    0: "On Time",
    1: "Late"
})
```
